```python
import pymc3 as pm
import matplotlib.pyplot as plt
import sklearn
from sklearn import datasets
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from sklearn.datasets import make_classification
from sklearn.datasets import make_moons
from sklearn.neural_network import MLPClassifier
import numpy as np
import theano
import theano.tensor as T
import warnings
warnings.filterwarnings('ignore')
floatX = theano.config.floatX
```

```python
# 4 class dataset
X, Y = make_classification(n_samples=1000, n_features=2,n_redundant=0,
                                     n_classes=4, n_clusters_per_class=1, random_state=0,
                                     weights=None, class_sep=1.5)
X = scale(X)
X = X.astype(floatX)
Y = Y.astype(floatX)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.5)

# plot dataset
_, ax = plt.subplots()
ax.scatter(X[Y==0, 0], X[Y==0, 1], label='Class 0')
ax.scatter(X[Y==1, 0], X[Y==1, 1], color='r', label='Class 1')
ax.scatter(X[Y==2, 0], X[Y==2, 1], color='p', label='Class 2')
ax.scatter(X[Y==3, 0], X[Y==3, 1], color='o', label='Class 3')

ax.legend()
ax.set(xlabel='X', ylabel='Y', title='Toy binary classification data set');
```

```python
# 4 class classic MLP
clf2 = MLPClassifier(solver='adam', hidden_layer_sizes=(5, 5, 5),activation='tanh', random_state=1)
clf2.fit(X_train, Y_train)
acc = clf2.score(X_test,Y_test)
pred2 = clf2.predict(X_test)


# plot predictions
_, ax = plt.subplots()
ax.scatter(X_test[pred2==0, 0], X_test[pred2==0, 1])
ax.scatter(X_test[pred2==1, 0], X_test[pred2==1, 1], color='r')
ax.scatter(X_test[pred2==2, 0], X_test[pred2==2, 1], color='g')
ax.scatter(X_test[pred2==3, 0], X_test[pred2==3, 1], color='y')
ax.set(title='Predicted labels in testing set', xlabel='X', ylabel='Y');
print('Accuracy = {}%'.format(acc * 100))
```

```python
def construct_nn(ann_input, ann_output):
    n_hidden = 5

    # Initialize random weights between each layer
    init_1 = np.random.randn(X.shape[1], n_hidden).astype(floatX)
    init_2 = np.random.randn(n_hidden, n_hidden).astype(floatX)
    init_out = np.random.randn(n_hidden,4).astype(floatX)

    with pm.Model() as neural_network:
        
        ann_input = pm.Data('ann_input', X_train)
        ann_output = pm.Data('ann_output', Y_train)

        # Weights from input to hidden layer
        weights_in_1 = pm.Normal('w_in_1', 0, sigma=1,
                                 shape=(X.shape[1], n_hidden),
                                 testval=init_1)

        # Weights from 1st to 2nd layer
        weights_1_2 = pm.Normal('w_1_2', 0, sigma=1,
                                shape=(n_hidden, n_hidden),
                                testval=init_2)

        # Weights from hidden layer to output
        weights_2_out = pm.Normal('w_2_out', 0, sigma=1,
                                  shape=(n_hidden,4),
                                  testval=init_out)

        # Build neural-network using tanh activation function
        act_1 = pm.math.tanh(pm.math.dot(ann_input,
                                         weights_in_1))
        act_2 = pm.math.tanh(pm.math.dot(act_1,
                                         weights_1_2))
        
        # softmax for output layer
        act_out = T.nnet.softmax(T.dot(act_2,
                                       weights_2_out))

        # Multiclass classification -> Categorical
        out = pm.Categorical('out',p=act_out,observed=ann_output,total_size=Y_train.shape[0])
    return neural_network

neural_network = construct_nn(X_train, Y_train)
```

```python
# random seed
from pymc3.theanof import set_tt_rng, MRG_RandomStreams
set_tt_rng(MRG_RandomStreams(42))

# train bayesian
with neural_network:
    inference = pm.ADVI()
    approx = pm.fit(n=10000, method=inference)
    
trace = approx.sample(draws=5000)
```

```python
# plot inference histogram 

plt.plot(approx.hist, label='ADVI', alpha=.3)
plt.legend()
plt.ylabel('ELBO')
plt.xlabel('iteration');
```

```python
# create symbolic input
x = T.matrix('X')
# symbolic number of samples is supported, we build vectorized posterior on the fly
n = T.iscalar('n')
x.tag.test_value = np.empty_like(X_train[:10])
n.tag.test_value = 100
_sample_proba = approx.sample_node(neural_network.out.distribution.p,
                                   size=n,
                                   more_replacements={neural_network['ann_input']: x})
# sampling
sample_proba = theano.function([x, n], _sample_proba)
```

```python
pred = np.argmax(sample_proba(X_test, 500).mean(0),axis=1)

_, ax = plt.subplots()
ax.scatter(X_test[pred==0, 0], X_test[pred==0, 1])
ax.scatter(X_test[pred==1, 0], X_test[pred==1, 1], color='r')
ax.scatter(X_test[pred==2, 0], X_test[pred==2, 1], color='g')
ax.scatter(X_test[pred==3, 0], X_test[pred==3, 1], color='y')
ax.set(title='Predicted labels in testing set', xlabel='X', ylabel='Y');
print('Accuracy = {}%'.format((Y_test == pred).mean() * 100))
```
